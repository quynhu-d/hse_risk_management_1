# Backtesting

```python
data = np.load('../results/simulations/mean_2022_portfolio_sim.npy')
```

```python
bt_data = pd.DataFrame(data[:,0].T).pct_change()
bt_data.index = pd.date_range(start = '2022-01-01', end = '2022-12-31', freq = 'd')
bt_data.columns = ['currencies', 'stocks', 'bonds']
bt_data['portfolio'] = bt_data['bonds'] * 50/260 + bt_data['stocks'] * 10/260 + \
bt_data['currencies'] * 200/260
```

```python
from var import VaR, load_data

var = {}
bth = {}
for pf in bt_data.columns:
    var[pf] = VaR(pd.DataFrame(bt_data.dropna()[pf]), weights=np.array([1]))
    var[pf].monte_carlo()
    try:
        bth[pf] = var[pf].backtest(method='h')
        var[pf].evaluate(bth[pf], )
    except:
        pass
```

```python
var['portfolio'].var_plot(bth['portfolio'])
```

```python
var['bonds'].var_plot(bth['bonds'])
```

```python
var['stocks'].var_plot(bth['stocks'])
```

```python
var['currencies'].var_plot(bth['currencies'])
```

# Real data

```python
real_data = pd.read_excel('../data/Final_data.xlsx')
real_data = real_data.fillna(method='ffill').fillna(method='bfill')
real_data = real_data.drop(real_data.tail(29).index)
real_data.set_index('Date', inplace=True)
# real_data.drop(columns='Date').pct_change().dropna().describe()
```

```python
currencies = ['Курс евро', 'Курс доллара']
stocks = [
    'Аэрофлот', 'Газпром', 'ГМК Норникель', 'Лукойл', 'МТС', 
    'ПИК', 'Роснефть', 'Сбербанк', 'Татнефть', 'Яндекс'
]
bonds = ['ОФЗ_26207', 'ОФЗ_26215', 'ОФЗ_26218', 'ОФЗ_26219', 'ОФЗ_26222']
# var_data = pd.DataFrame(index=real_data.Date)
var_data = pd.DataFrame()
var_data['currencies'] = real_data.pct_change().dropna()[currencies].mean(axis=1)
var_data['stocks'] = real_data.pct_change().dropna()[stocks].mean(axis=1)
var_data['bonds'] = real_data.pct_change().dropna()[bonds].mean(axis=1)
# var_data = var_data.pct_change().dropna()
var_data.index = real_data.index[1:]
var_data['portfolio'] = var_data['bonds'] * 50/260 + var_data['stocks'] * 10/260 + \
var_data['currencies'] * 200/260
var_data.tail(2)  # returns
```

```python
## parametric
param_ve = Parametric_VE(var_data.portfolio)
param_ve.get_var_es()
param_ve.plot_dist()
```

```python
## historical
hist_ve = Historical_VE(var_data.portfolio)
hist_ve.get_var_es()
hist_ve.plot_dist()
```

```python
## bootstrap
boot_ve = Bootstrap_VE(var_data.portfolio)
boot_ve.get_var_es()
boot_ve.plot_dist()
```

```python
from tabulate import tabulate

print(
    tabulate(
        [
            ['Parametric, 1d', param_ve.var99_1d, param_ve.cvar975_1d], 
            ['Historical, 1d', hist_ve.var99_1d, hist_ve.cvar975_1d],
            ['Parametric, 10d', param_ve.var99_10d, param_ve.cvar975_10d], 
            ['Historical, 10d', hist_ve.var99_10d, hist_ve.cvar975_10d],
            ['Bootstrapping, 1d', boot_ve.var99_1d, boot_ve.cvar975_1d], 
            ['Bootstrapping, 10d', boot_ve.var99_10d, boot_ve.cvar975_10d]
        ],
        headers=['Approach', 'Value at Risk 99%', 'Conditional Value at Risk (ES) 97.5%'],
        tablefmt = 'fancy_grid',
        stralign = 'center',
        numalign= 'center',
        floatfmt=".4f"
    )
)
```

```python
## backtesting
var = VaR(pd.DataFrame(var_data.portfolio), weights=np.array([1]))
try:
    bth = var.backtest(method='h')
    var.evaluate(bth)
except:
    pass
```

```python
var.var_plot(bth)
```
